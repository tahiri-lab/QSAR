```python
import pandas as pd
```

```python
# Global variable
TRAINING: str = "TRAINING"
TEST: str = "TEST"


# Modifying the values
SHOULD_FUSE_PRED: bool = True # Fuse the column PRED_neutral (neutral) with $PRED_516 (ionizable)
```

# Opening data and comparing data

## Ionizable

```python
df_ionizable: pd.DataFrame = pd.read_csv("../Data/ionizable_dataset_72.csv", delimiter=';')
df_ionizable.drop(["index"], axis=1, inplace=True)
df_ionizable_train: pd.DataFrame = df_ionizable.loc[df_ionizable["set"] == TRAINING]
df_ionizable_test: pd.DataFrame = df_ionizable.loc[df_ionizable["set"] == TEST]
display(df_ionizable.head())
print("===== ALL =====")
display(df_ionizable.shape)
print("===== TRAIN =====")
display(df_ionizable_train.shape)
print("===== TEST =====")
display(df_ionizable_test.shape)
```

## Neutral

```python
df_neutral: pd.DataFrame = pd.read_csv("../Data/neutral_dataset_111.csv", delimiter=';')
df_neutral.drop(["index"], axis=1, inplace=True)
df_neutral_train: pd.DataFrame = df_neutral.loc[df_neutral["set"] == TRAINING]
df_neutral_test: pd.DataFrame = df_neutral.loc[df_neutral["set"] == TEST]
display(df_neutral.head())
print("===== ALL =====")
display(df_neutral.shape)
print("===== TRAIN =====")
display(df_neutral_train.shape)
print("===== TEST =====")
display(df_neutral_test.shape)
```

## All

### Comparing ALL (test) with ionizable and neutral before merge
Comparison done to check the header differences between all dataframe and see if some need to be merged together

```python
df_all: pd.DataFrame = pd.read_csv("../Data/full_dataset_test.csv", delimiter=';')


# Difference of headers in the dataframe
print("===== HEADER COMPARISON =====")
print(">>> NEUTRAL VS ALL")
display(df_neutral.columns.difference(df_all.columns))
print(">>> ALL VS NEUTRAL")
display(df_all.columns.difference(df_neutral.columns))

print(">>> IONIZABLE VS ALL")
display(df_ionizable.columns.difference(df_all.columns))
print(">>> ALL VS IONIZABLE")
display(df_all.columns.difference(df_ionizable.columns))

print(">>> NEUTRAL VS IONIZABLE")
display(df_neutral.columns.difference(df_ionizable.columns))

print(">>> IONIZABLE VS NEUTRAL")
display(df_ionizable.columns.difference(df_neutral.columns))
```

## Merging all of them together

```python
df_buffer_neutral = df_neutral
df_buffer_ionizable = df_ionizable

if SHOULD_FUSE_PRED:
    df_buffer_neutral.rename(columns={"$PRED_neutral": "$PRED"}, inplace=True)
    df_buffer_ionizable.rename(columns={"$PRED_516": "$PRED"}, inplace=True)


print(df_all.shape)
df_all = pd.concat([df_all, df_buffer_neutral], ignore_index=True)
print(df_all.shape)
df_all = pd.concat([df_all, df_buffer_ionizable], ignore_index=True)
print(df_all.shape)
df_buffer_neutral
```
