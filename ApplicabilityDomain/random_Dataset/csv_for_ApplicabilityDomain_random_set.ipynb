## Preparing the input file for the Applicability Domain random split
We need 2 csv files : taining and test set with: 
> First Row: Header names \
> First column: Chemical serial number (index) \
> Subsequent columns: Descriptors values \


```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Import data to pandas dataframe and set the first column as index
df = pd.read_excel ('milk_plasma_all_data_mrg_020821.xlsx')
df.set_index('index', inplace=True)
#number of rows, number of columns
df.shape
```

```python
# need more info about 'group' column! how we got 0 , 1, 2 ?
# the split was made according to the charge? 'set' column

# I will keep the columns 'neutral', 'CHARGED'
```

```python
#   * Drop the columns : 'smiles','conc','name','CLASS','MP_RATIO','Log_MP_RATIO', 'neutral','TEST','CHARGED


new_df=df.drop(columns=['smiles','conc','name','CLASS','MP_RATIO','Log_MP_RATIO','set','TEST'])

new_df
```

```python
#list of all the column names  containing '$' in the first position
cols = [c for c in df.columns if c[0] == '$']

print(cols)
```

```python
#   * Drop the columns starting with $

new_df=new_df.drop(cols, axis=1)
new_df
```

```python
print(new_df.dtypes)
print('______________________________________')
for col in new_df.columns :
    if df[col].dtypes == 'object':
        print(col,'<- is object type')
```

```python
new_df
```

```python
new_df=new_df.drop(columns=['FP:MACCS'])
new_df
```

```python
##Slicing the data set into training and test set
#The random_state is a pseudo-random number parameter that reproduce 
#the same exact train test split each time the code is run 

X_train, X_test = train_test_split(new_df, random_state=0, 
                                   train_size = .80)
X_train.shape, X_test.shape
```

```python
X_train
```

```python
#Exporting the test and taining dataFrames to CSVs
X_train.to_csv('milk_plasma_data_trainingSet_random.csv', encoding='utf-8')
X_test.to_csv('milk_plasma_data_testSet_random.csv', encoding='utf-8')
```
