## Preparing the input file for the Applicability Domain
We need 2 csv files : taining and test set with: 
> First Row: Header names \
> First column: Chemical serial number (index) \
> Subsequent columns: Descriptors values \


```python
import pandas as pd
```

```python
#Import data to pandas dataframe and set the first column as index
df = pd.read_excel ('milk_plasma_all_data_mrg_020821.xlsx')
df.set_index('index', inplace=True)
#number of rows, number of columns
df.shape
```

```python
# need more info about 'group' column! how we got 0 , 1, 2 ?
# the split was made according to the charge? 'set' column

# I will keep the columns 'neutral', 'CHARGED' and 'TEST'
```

```python
#   * Drop the columns : 'smiles','conc','name','CLASS','MP_RATIO','Log_MP_RATIO', 'neutral','TEST','CHARGED


new_df=df.drop(columns=['smiles','conc','name','CLASS','MP_RATIO','Log_MP_RATIO','set'])

new_df
```

```python
#list of all the column names  containing '$' in the first position
cols = [c for c in df.columns if c[0] == '$']

print(cols)
```

```python
#   * Drop the columns starting with $

new_df=new_df.drop(cols, axis=1)
new_df
```

```python
print(new_df.dtypes)
print('______________________________________')
for col in new_df.columns :
    if df[col].dtypes == 'object':
        print(col,'<- is object type')
```

```python
new_df
```

```python
new_df=new_df.drop(columns=['FP:MACCS'])
new_df
```

```python
#split the dataframe according to the column (RZ) -> 'TEST' 
#as defined in the 'milk_plasma_all_data_mrg_020821.xlsx' first sheet 'full_dataset'

split = new_df.groupby('TEST')
test = split.get_group(0)
training = split.get_group(1)

df_Test = test.apply(lambda x: x)
df_Training = training.apply(lambda x: x)

print('df_Test.shape => ',df_Test.shape)
print('df_Training.shape => ',df_Training.shape)
```

```python
#Drop the TEST column 
df_Test = df_Test.drop(columns=['TEST'])
df_Training = df_Training.drop(columns=['TEST'])
```

```python
#Exporting the test and taining dataFrames to CSVs
df_Training.to_csv('milk_plasma_data_trainingSet_selected.csv', encoding='utf-8')
df_Test.to_csv('milk_plasma_data_testSet_selected.csv', encoding='utf-8')
```
